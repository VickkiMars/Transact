```python
import random
```

```python
def one_zero(number:str):
	newf = ""
	number = str(number)
	for i in number:
		if i != number[-1]:
			newf += i
		else:
			newf += "0"
			break
	return newf


def two_zero(number:str):
	newf = ""
	number = str(number)
	try:
		for i in number:
			if i != number[-2]:
				newf += i
			else:
				newf += "00"
				break
	except Exception as e:
		return None
	else: return newf

def three_zero(number:str):
	newf = ""
	number = str(number)
	try:
		for i in number:
			if i != number[-3]:
				newf += i
			else:
				newf += "000"
				break
	except Exception as e:
		return None
	else:
		return newf
```

```python
def amounts():
	upper_limits = [1000, 10000, 100000, 1000000,1000000000 ]
	lower_limits = [50, 1000, 10000, 100000, 1000000]
	ranger = [0,1,2,3,4]
	ch = random.choice(ranger)
	amount = random.randint(lower_limits[ch], upper_limits[ch])
	print(amount)
	choice_list = [amount, one_zero(amount), two_zero(amount), three_zero(amount)]
	for i in choice_list:
		if i == None:
			choice_list.remove(i)

	amount_ch = random.choice(choice_list)
	amount  = amount_ch if int(amount_ch) != 0 else amount
	amount = str(amount)
	fin_suffix = {"k":1000, "m":1000000, "h":100, "b":1000000000, "t":1000000000000}
	if len(amount) < 4:
		keychoice = "h"
	elif len(amount) >= 4 and len(amount) < 7:
		keychoice = "k"
	elif len(amount) > 6 and len(amount) < 10:
		keychoice = "m"
	elif len(amount) > 8 and len(amount < 13):
		keychoice = "b"
	else:
		keychoice = "t"

	amount = random.choice([amount, f"{int(int(amount)/fin_suffix[keychoice])}{keychoice}"])

	return amount
```

```python
def load_and_split_file(filename):

	with open(filename, 'r') as file:
		content = file.read()
	lines = content.split('\n')
	return lines


# Example usage
filename = 'C:/Users/Victor/Desktop/Transaction/.txt files/banks.txt'  # Replace with the path to your file
lines = load_and_split_file(filename)

accountnames = load_and_split_file(r"C:/Users/Victor/Desktop/Transaction/.txt files/names.txt")
```

```python
fin_suffix = {"k":1000, "m":1000000, "h":100, "b":1000000000, "t":1000000000000}
```

```python
with open(r"C:\Users\Victor\Desktop\Transaction\.txt files\dataset.txt", "a") as dataset:
	dataset.write(f"Input Data\tOutput Data")
	for i in range(200000):
		choice = random.choice(["T", "N", "W", "V"])
		account_number = random.randint(1000000000, 9999999999)
		keyword = random.choice(keywords)
		bank_name = random.choice(lines)
		account_name = random.choice(accountnames)
		if len(account_name) < 10:
			continue
		amount = amounts() #970k

		for suffix in fin_suffix.keys():
			if suffix in str(amount):
				lamount = int(str(amount).split(suffix)[0]) * fin_suffix[suffix] #970000
				break
			else:
				lamount = amount


		if choice != "N":
			input_format = f"{keyword} {amount} to {account_number} {bank_name}"	
			output_format =  f"{lamount} | {account_number} | {bank_name}"
			dataset.write(f"{input_format}\t{output_format}\n")
		elif choice == "W":
			input_format = f"{keyword} {amount} to {bank_name} {account_name}"
			output_format =  f"{lamount} | {bank_name} | {account_name}"
			dataset.write(f"{input_format}\t{output_format}\n")
		elif choice == "V":
			input_format = f"{amount} to {bank_name} {account_name}"
			output_format =  f"{lamount} | {bank_name} | {account_name}"
			dataset.write(f"{input_format}\t{output_format}\n")
		else:
			input_format = f"{keyword} {amount} to {account_number} {account_name}"
			output_format =  f"{lamount} | {account_number} | {account_name}"
			dataset.write(f"{input_format}\t{output_format}\n")

	perc = i/1000
	if perc % 5 == 0:
		print(f"{perc}%")
		
```
